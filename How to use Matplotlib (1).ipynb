```python
#Importing libraries
```

```python
import matplotlib.pyplot as plt
```

```python
#Lineal Plot
```

```python
plt.figure(figsize=(10,7))

#Pais 1
x = [2016,2017,2018,2019,2020,2021]
y = [45, 46, 47, 48, 50, 51]


#Pais 2
x2= [2016,2017,2018,2019,2020,2021]
y2= [47,46,47, 49, 52, 53]


plt.plot(x,y,marker = "o", linestyle="--", color = "y", label="pais 1")
plt.plot(x2, y2, marker = "o", linestyle="--", color = "b", label="pais 2")

plt.xlabel("A침os")
plt.ylabel("Poblaci칩n ( en millones)")
plt.title("Poblaci칩n por a침o")
plt.yticks([46,48,51])
plt.legend(loc="lower right")
plt.savefig("ejemplo.png")
plt.show()
```

```python
# Subplot
```

```python
fig, ax = plt.subplots(1,2, sharey=True)
ax[0].plot(x,y, color="g", label="pais 1")
ax[1].plot(x2,y2,color="b", label="pais 2")
plt.show()
```
